```python
import json

with open("./travel/02_theme_codes_llm.txt", "r") as f:
    content = f.read().strip().split("\n\n")
    documents = [json.loads(p) for p in content]


for i, document in enumerate(documents[:]):
    print(f'= File {i+1} =')
    for i, key in enumerate(document.keys()):
        if key == "file":
            continue
        print(f'{i+1}. {key}')
        for j, item in enumerate(document[key]):
            print(f"{i+1}.{j+1}. {item['code']}")
            for quote in item["quotes"]:
                print(f" - {quote}")
        print()
```
