# Template

```python
import pandas as pd
import plotly.express as px
import os

df = pd.read_csv('./NetLogo Model/outputs.csv')

# Create line chart using plotly
fig = px.line(df, title='Mode of Transport', labels={'value':'Number of People', 'index':'Time Step', 'variable':'Mode'})

# Update layout for better readability
fig.update_layout(title_x=0.5,  # Center the title
                  xaxis_title='Time Step',
                  yaxis_title='Number of People',
                  legend_title='Mode of Transport')

# Save visualisation image in folder visualisations
os.makedirs("results_2/visualisations", exist_ok=True)
fig.write_image("results_2/visualisations/transport_modes.png")

#Optionally display the figure
#fig.show()
```

```python
import pandas as pd
import plotly.express as px
import os

df = pd.read_csv('./NetLogo Model/outputs_25.csv')

# Create line chart using plotly
fig = px.area(df, title='Mode of Transport (25% environment concious commuter)', labels={'value':'Number of People', 'index':'Time Step', 'variable':'Mode'})

# Update layout for better readability
fig.update_layout(title_x=0.5,  # Center the title
                  xaxis_title='Time Step',
                  yaxis_title='Number of People',
                  legend_title='Mode of Transport')

# Save visualisation image in folder visualisations
os.makedirs("results_2/visualisations", exist_ok=True)
fig.write_image("results_2/visualisations/v25_fill.png")

#Optionally display the figure
fig.show()
```
